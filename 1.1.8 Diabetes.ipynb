### Linear Regression for Diabetes 

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error 
from sklearn.model_selection import train_test_split
```

```python
diabetes= datasets.load_diabetes()  # Load the diabetes dataset
X=diabetes.data                     # Indepenent variables
y=diabetes.target                   # Depenent variables
```

```python
Xtrain, Xtest, ytrain, ytest=train_test_split(X,y,test_size=0.5, random_state=0)
```

```python
# Fit a Linear Regression model
regr=linear_model.LinearRegression()
# Train the model using the training sets 
regr.fit(Xtrain, ytrain)
# Make prediction using the test set
ypred=regr.predict(Xtest)
```

```python
# The coefficients 
print('Coefficients: \n', regr.coef_)

# Intercept
print('Intercept: \n', regr.intercept_)
```

```python
# Explained variance score: 1 is perfect prediction
print("R^2=%5.3f" % r2_score(ytest,ypred))

# The mean absolute error 
print("MAE=%5.3f" % mean_absolute_error(ytest,ypred))

# The mean sqaure error 
print("MAE=%5.3f" % mean_squared_error(ytest,ypred))
```

```python
# Plot outputs 
fig, ax=plt.subplots()
ax.scatter(ytest,ypred,edgecolors=(0,0,0))
ax.plot([y.min(),y.max()],[y.min(),y.max()], 'k--',lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
